## 3. Quantity Validation

The 'Quantity' column was examined to identify zero, negative, and unusually large values.

Negative quantities may represent product returns, cacellations, or inventory adjustments rather than data entry errors.

```python
# Summary statistics for Quantity
df["Quantity"].describe()
```

```python
# Count zero and negative quantity records
zero_qty_count = (df["Quantity"]==0).sum()
negative_qty_count = (df["Quantity"]<0).sum()
positive_qty_count = (df["Quantity"]>0).sum()

print(f"Zero_quantity records    : {zero_qty_count:,}")
print(f"Negative_quantity records: {negative_qty_count:,}")
print(f"Positive quantity records: {positive_qty_count:,}")
```

```python
# Preview negative quantity records
negative_qty = df[df["Quantity"] <0]

negative_qty.head(10)
```

### Interpretation

A total of 19,863 records contained negative quantities.

Further investigation showed that these transactions had invoice numbers beginning with the letter **C**, indicating cancelled or returned orders.

However, a small number of negative quantity records did not follow this pattern. These records will be investigated separately before making any cleaning decisions. 

At this stage, no negative quantity records are removed.

```python
# Verify whether all negative quantities are cancellation invoices
negative_qty["Invoice"].astype(str).str.startswith("C").value_counts()
```

```python
# Investigate negative quantities without cancellation invoice
negative_without_c = negative_qty[
    ~negative_qty["Invoice"].astype(str).str.startswith("C")
]

negative_without_c.head(20)
```

```python
negative_without_c.shape
```

### Business Interpretation

Negative quantity records were devided into two categories:

- Customer return or calcellation transactions (Invoice begins with **C**)
- Interanl inventory adjustment records (e.g., *damaged*, *lost*, *short*, *bad quality*)

Return transactions were retained becaue they represent valid business events.

Inventory adjustment records were removed because they do not represent customer purchases and are outside the scope of this sales analysis project.

```python
inventory_adjustment = negative_without_c

print("Inventory_adjustment records:", len(inventory_adjustment))
```

```python
df = df.drop(inventory_adjustment.index)
```

```python
print(df.shape)
print(
    df[
        (df["Quantity"] < 0) &
        (~df["Invoice"].astype(str).str.startswith("C"))
    ].shape
)
```

```python
inventory_adjustment = negative_without_c

#Save for audit
inventory_adjustment.to_csv(
    "../data/processed/inventory_adjustments.csv",
    index=False
)
```
